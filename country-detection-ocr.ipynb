```python
import os
import glob
import pandas as pd

def create_file_country_df(base_path="/kaggle/input/passports/dataset"):
    data = []
    
    if not os.path.exists(base_path):
        print(f"Путь не найден: {base_path}")
        return pd.DataFrame(columns=["photo", "code"])
    
    countries = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]
    
    for country in countries:
        country_path = os.path.join(base_path, country)
        files = glob.glob(os.path.join(country_path, "*.jpg"))
        
        for file_path in files:
            file_name = os.path.basename(file_path)
            data.append({"photo": base_path + '/' +country + '/' +file_name, "code": country})
    
    df = pd.DataFrame(data)
    return df

# Использование
df = create_file_country_df()
df.head()
```

```python
import cv2
import numpy as np
import pandas as pd
from pathlib import Path
import random
import os
from sklearn.model_selection import train_test_split

def augment_image(image_path: str, output_dir: str, base_name: str, code: str, num_augmentations: int = 2):
    """
    Создает num_augmentations аугментированных версий изображения
    """
    img = cv2.imread(image_path)
    if img is None:
        return []
    
    h, w = img.shape[:2]
    augmented_paths = []
    
    for i in range(num_augmentations):
        # Применяем случайные аугментации
        aug_img = img.copy()
        
        # 1. Геометрические трансформации (легкие для OCR)
        if random.random() < 0.7:  # 70% вероятность
            aug_img = apply_geometric_transform(aug_img)
        
        # 2. Изменения яркости и контраста
        if random.random() < 0.8:  # 80% вероятность
            aug_img = adjust_brightness_contrast(aug_img)
        
        # 3. Размытие (осторожно для текста)
        if random.random() < 0.3:  # 30% вероятность - слабое размытие
            aug_img = apply_blur(aug_img)
        
        # 4. Шум
        if random.random() < 0.4:  # 40% вероятность
            aug_img = add_noise(aug_img)
        
        # 5. Цветовые искажения
        if random.random() < 0.5:  # 50% вероятность
            aug_img = color_distortion(aug_img)
        
        # 6. Компрессия (имитация JPEG артефактов)
        if random.random() < 0.3:  # 30% вероятность
            aug_img = jpeg_compression(aug_img)
        
        # Сохраняем аугментированное изображение
        aug_path = os.path.join(output_dir, f"{base_name}_aug_{i}.jpg")
        cv2.imwrite(aug_path, aug_img, [cv2.IMWRITE_JPEG_QUALITY, 95])
        augmented_paths.append(aug_path)
    
    return augmented_paths

def apply_geometric_transform(img):
    """Применяет случайные геометрические трансформации"""
    h, w = img.shape[:2]
    
    # Случайный выбор типа трансформации
    transform_type = random.choice(['rotate', 'affine', 'perspective'])
    
    if transform_type == 'rotate':
        # Поворот на небольшой угол (-3 до +3 градусов)
        angle = random.uniform(-3, 3)
        center = (w // 2, h // 2)
        M = cv2.getRotationMatrix2D(center, angle, 1.0)
        return cv2.warpAffine(img, M, (w, h), borderValue=(255, 255, 255))
    
    elif transform_type == 'affine':
        # Легкая аффинная трансформация
        pts1 = np.float32([[0, 0], [w-1, 0], [0, h-1]])
        # Небольшие смещения углов
        shift = min(w, h) * 0.02  # 2% от размера
        pts2 = np.float32([
            [random.uniform(-shift, shift), random.uniform(-shift, shift)],
            [w-1 + random.uniform(-shift, shift), random.uniform(-shift, shift)],
            [random.uniform(-shift, shift), h-1 + random.uniform(-shift, shift)]
        ])
        M = cv2.getAffineTransform(pts1, pts2)
        return cv2.warpAffine(img, M, (w, h), borderValue=(255, 255, 255))
    
    else:  # perspective
        # Легкая перспективная трансформация
        shift = min(w, h) * 0.03  # 3% от размера
        pts1 = np.float32([[0, 0], [w-1, 0], [w-1, h-1], [0, h-1]])
        pts2 = np.float32([
            [random.uniform(-shift, shift), random.uniform(-shift, shift)],
            [w-1 + random.uniform(-shift, shift), random.uniform(-shift, shift)],
            [w-1 + random.uniform(-shift, shift), h-1 + random.uniform(-shift, shift)],
            [random.uniform(-shift, shift), h-1 + random.uniform(-shift, shift)]
        ])
        M = cv2.getPerspectiveTransform(pts1, pts2)
        return cv2.warpPerspective(img, M, (w, h), borderValue=(255, 255, 255))

def adjust_brightness_contrast(img):
    """Изменяет яркость и контраст"""
    # Случайные параметры в безопасном диапазоне для OCR
    alpha = random.uniform(0.8, 1.3)  # контраст
    beta = random.uniform(-30, 30)    # яркость
    
    return cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

def apply_blur(img):
    """Применяет легкое размытие"""
    blur_type = random.choice(['gaussian', 'motion'])
    
    if blur_type == 'gaussian':
        # Легкое гауссово размытие
        kernel_size = random.choice([3, 5])
        return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)
    else:
        # Motion blur (имитация движения камеры)
        size = random.choice([3, 5])
        kernel = np.zeros((size, size))
        kernel[int((size-1)/2), :] = np.ones(size)
        kernel = kernel / size
        return cv2.filter2D(img, -1, kernel)

def add_noise(img):
    """Добавляет шум"""
    noise_type = random.choice(['gaussian', 'salt_pepper'])
    
    if noise_type == 'gaussian':
        # Гауссов шум
        mean = 0
        std = random.uniform(5, 15)
        noise = np.random.normal(mean, std, img.shape).astype(np.uint8)
        return cv2.add(img, noise)
    else:
        # Salt and pepper noise
        noise = np.random.random(img.shape[:2])
        img_noisy = img.copy()
        img_noisy[noise < 0.01] = 0      # pepper
        img_noisy[noise > 0.99] = 255    # salt
        return img_noisy

def color_distortion(img):
    """Цветовые искажения"""
    # Конвертируем в HSV для работы с цветом
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Случайные изменения
    hue_shift = random.uniform(-10, 10)
    sat_scale = random.uniform(0.8, 1.2)
    val_scale = random.uniform(0.9, 1.1)
    
    hsv[:, :, 0] = np.clip(hsv[:, :, 0] + hue_shift, 0, 179)
    hsv[:, :, 1] = np.clip(hsv[:, :, 1] * sat_scale, 0, 255)
    hsv[:, :, 2] = np.clip(hsv[:, :, 2] * val_scale, 0, 255)
    
    return cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2BGR)

def jpeg_compression(img):
    """Имитирует JPEG компрессию"""
    quality = random.randint(70, 95)
    encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), quality]
    _, encimg = cv2.imencode('.jpg', img, encode_param)
    return cv2.imdecode(encimg, 1)

# Основная функция для создания расширенного датасета
def create_augmented_dataset(df: pd.DataFrame, output_dir: str, augmentations_per_image: int = 2):
    """
    Создает аугментированный датасет
    df: исходный DataFrame с колонками ['photo', 'code']
    output_dir: папка для сохранения аугментированных изображений
    augmentations_per_image: количество аугментаций на каждое изображение
    """
    
    os.makedirs(output_dir, exist_ok=True)
    
    augmented_data = []
    
    # Добавляем оригинальные изображения
    for _, row in df.iterrows():
        augmented_data.append({
            'photo': row['photo'],
            'code': row['code'],
            'is_augmented': False
        })
    
    # Создаем аугментации
    print(f"Создаем аугментации для {len(df)} изображений...")
    
    for idx, row in df.iterrows():
        image_path = row['photo']
        code = row['code']
        
        # Получаем базовое имя файла
        base_name = Path(image_path).stem
        
        # Создаем аугментации
        aug_paths = augment_image(
            image_path, 
            output_dir, 
            f"{code}_{base_name}", 
            code, 
            augmentations_per_image
        )
        
        # Добавляем пути к аугментированным изображениям
        for aug_path in aug_paths:
            augmented_data.append({
                'photo': aug_path,
                'code': code,
                'is_augmented': True
            })
        
        if (idx + 1) % 50 == 0:
            print(f"Обработано {idx + 1}/{len(df)} изображений")
    
    # Создаем новый DataFrame
    augmented_df = pd.DataFrame(augmented_data)
    
    print(f"Исходный датасет: {len(df)} изображений")
    print(f"Расширенный датасет: {len(augmented_df)} изображений")
    print(f"Коэффициент увеличения: {len(augmented_df) / len(df):.1f}x")
    
    return augmented_df

# Использование с вашим DataFrame
augmented_df = create_augmented_dataset(df, '/kaggle/working/augmented_images', augmentations_per_image=2)

# Train/test split для расширенного датасета
train_df_aug, test_df_aug = train_test_split(
    augmented_df, 
    test_size=0.2, 
    random_state=42,
    stratify=augmented_df['code']
)

print(f"Обучающий набор: {len(train_df_aug)} изображений")
print(f"Тестовый набор: {len(test_df_aug)} изображений")
```

```python
import os, shutil
from pathlib import Path

# Входные данные
all_dir  = Path("/kaggle/working/test_images_all")
orig_dir = Path("/kaggle/working/test_images_orig")

all_dir.mkdir(parents=True, exist_ok=True)
orig_dir.mkdir(parents=True, exist_ok=True)

def safe_copy(src_path, dst_dir):
    src = Path(src_path)
    if not src.is_file():
        return False, f"missing: {src}"
    # Разрешение коллизий имён
    dst = dst_dir / src.name
    if dst.exists():
        stem, suf = dst.stem, dst.suffix
        i = 1
        while dst.exists():
            dst = dst_dir / f"{stem}__{i}{suf}"
            i += 1
    shutil.copy2(src, dst)
    return True, str(dst)

copied_all, copied_orig, missing = 0, 0, []

# Копируем всё в all
for p in test_df_aug["photo"]:
    ok, info = safe_copy(p, all_dir)
    if ok: 
        copied_all += 1
    else:
        missing.append(info)

# Копируем только оригиналы (is_augmented == False) в orig
for p, is_aug in zip(test_df_aug["photo"], test_df_aug["is_augmented"]):
    if not bool(is_aug):
        ok, info = safe_copy(p, orig_dir)
        if ok: 
            copied_orig += 1
        else:
            missing.append(info)

print(f"All copied: {copied_all}, Orig copied: {copied_orig}")
if missing:
    print(f"Missing files: {len(missing)} (showing first 10)")
    print(missing[:10])
```

```python
import re, unicodedata, os
import pandas as pd
from pathlib import Path

# Регулярка для строк вида:
# [..] ppocr INFO: Predicts of /path/to/img.jpg:('BEL', 0.978121280670166)
_LINE_RE = re.compile(
    r"Predicts of\s+(?P<path>.+?):\((?P<tuple>.+)\)"
)

def _strip_path(p: str) -> str:
    # Обрезаем до имени файла для стабильного merge 
    return os.path.basename(str(p))

def _normalize_code(s: str) -> str:
    s = unicodedata.normalize("NFKC", s)
    s = re.sub(r"[^A-Za-z]", "", s).upper()
    return s

def parse_predict_rec_log_to_df(path_to_log: str, test_df_aug: pd.DataFrame) -> pd.DataFrame:
    """
    Возвращает DF со столбцами:
      image (basename), pred_text (как в логе), conf (float), pred_code (ISO‑3 нормализация)
    """
    rows = []
    with open(path_to_log, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            m = _LINE_RE.search(line)
            if not m:
                continue
            img_path = m.group("path").strip().strip('"').strip("'")
            tup = m.group("tuple").strip()
            # Ожидаем ('TEXT', 0.97). Безопасный парсинг:
            pred_text, conf = "", 0.0
            # Вытаскиваем первую подстроку в кавычках как текст
            m_txt = re.search(r"[\"']([^\"']+)[\"']", tup)
            if m_txt:
                pred_text = m_txt.group(1)
            # Вытаскиваем первое число как score
            m_num = re.search(r"([-+]?\d*\.\d+|\d+)(e[-+]?\d+)?", tup)
            if m_num:
                try:
                    conf = float(m_num.group(0))
                except:
                    conf = 0.0
            rows.append({
                "image": _strip_path(img_path),
                "pred_text": pred_text,
                "conf": conf
            })
    pred_df = pd.DataFrame(rows).drop_duplicates(subset=["image"], keep="last")

    # Сопоставляем с тестовой таблицей по basename
    ref = test_df_aug.copy()
    ref["image"] = ref["photo"].map(_strip_path)
    # Итог: image, true_code, pred_text, pred_code, conf
    out = ref[["image", "code"]].merge(pred_df, on="image", how="left")
    out = out.rename(columns={"code": "true_code"})
    out["pred_code"] = out["pred_text"].fillna("").map(_normalize_code)
    return out
```

```python
from sklearn.metrics import accuracy_score, classification_report

def eval_metrics(df_preds: pd.DataFrame, y_col="true_code", yhat_col="pred_code"):
    y_true = df_preds[y_col].astype(str).str.upper().fillna("UNKNOWN")
    y_pred = df_preds[yhat_col].astype(str).str.upper().fillna("UNKNOWN")
    acc = accuracy_score(y_true, y_pred)  # Общая точность. [web:1045]
    report = classification_report(y_true, y_pred, zero_division=0)  # По‑классовый отчёт. 
    return acc, report
```
